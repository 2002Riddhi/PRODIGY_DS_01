```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
# Define gender counts
gender_counts = titanic['Sex'].value_counts().reset_index()
gender_counts.columns = ['Sex', 'Count']

# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Bar chart for gender distribution
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Count', hue='Sex', data=gender_counts, palette='pastel', dodge=False, legend=False)
plt.title('Gender Distribution on Titanic', fontsize=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
for i, count in enumerate(gender_counts['Count']):
    plt.text(i, count + 10, str(count), ha='center', fontsize=12)
plt.show()
```

```python
# Histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution on Titanic', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
